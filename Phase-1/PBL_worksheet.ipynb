```python
import pandas as pd
import numpy as np
```

```python
#creating dataset 
np.random.seed(4)
n=1000
data={
    'Packet Size': np.random.randint(50,1500,n),
    'Connection Duration': np.random.randint(1,300,n),
    'Error Rate': np.random.random(n)
}
dataframe=pd.DataFrame(data)

#creating column 'label' as target
dataframe['Label']=((dataframe['Packet Size']>1000)|(dataframe['Error Rate']>0.6)).astype(int)
dataframe.head()
dataframe.to_csv("Network_Log.csv",index=False)
```

```python
#load dataset and separate features and label
from sklearn.preprocessing import StandardScaler
df=pd.read_csv("Network_Log.csv")
x=df.iloc[:,:-1]
y=df.iloc[:,-1]

#standarisation (feature scaling)
ss=StandardScaler()
x_scaled=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
#Train-test split
x_train, x_test, y_train, y_test= train_test_split(x_scaled, y, test_size=0.2, random_state=4)
```
