```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

#-----------data preprocessing------------
df=pd.read_csv("Processed_data.csv")
df
```

```python
x=df.drop('Label',axis=1)
y=df['Label']
#Train-test split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2, random_state=4)
```

```python
#feature scaling(Standarisation)
ss=StandardScaler()

x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
#training model
from sklearn.neighbors import KNeighborsClassifier

error_rates = []

# try different K values to find the optimal value of K
for k in range(1, 21, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    pred_k = knn.predict(x_test)
    
    error = (pred_k != y_test).mean()
    error_rates.append(error)
    
k_values = list(range(1, 21, 2))
best_k = k_values[error_rates.index(min(error_rates))]
print("Best K:", best_k,"\n")
# plot graph
plt.figure(figsize=(8,5))
plt.plot(range(1,21,2), error_rates, marker='o',color='green')
plt.xlabel("K value")
plt.ylabel("Error Rate")
plt.title("Elbow Method")
plt.show()
```

```python
#According to graph we are getting optimal value of K as 9
#model 1
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(x_train, y_train)
# Model training completed.
```

```python
#model 2
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
#model 3
dtc=DecisionTreeClassifier(random_state=10)
dtc.fit(x_train, y_train)
```

```python
#model 4
rfc=RandomForestClassifier(n_estimators=100,random_state=10)
rfc.fit(x_train, y_train)
```

```python
#evaluation of all models and finding the best one out of all
#1.Accuracy 
y_pred_knn=knn.predict(x_test)
print("KNN Accuracy:",accuracy_score(y_test,y_pred_knn))

y_pred_lr=lr.predict(x_test)
print("Logistic Regression Accuracy:",accuracy_score(y_test,y_pred_lr))

y_pred_dtc=dtc.predict(x_test)
print("Decision Tree Classifier Accuracy:",accuracy_score(y_test,y_pred_dtc))

y_pred_rfc=rfc.predict(x_test)
print("Random Forest Classifier Accuracy:",accuracy_score(y_test,y_pred_rfc))
```

```python
#2.Confusion Matrix
print("KNN Confusion matrix:\n",
      confusion_matrix(y_test,y_pred_knn))

print("Logistic Regression Confusion matrix:\n",
      confusion_matrix(y_test,y_pred_lr))

print("Decision Tree Classifier Confusion matrix:\n",
      confusion_matrix(y_test,y_pred_dtc))

print("Random Forest Classifier Confusion matrix:\n",
      confusion_matrix(y_test,y_pred_rfc))
```

```python
#3.Classification report
print("KNN Classification Report:\n",
      classification_report(y_test,y_pred_knn))

print("Logistic Regression Classification Report:\n",
      classification_report(y_test,y_pred_lr))

print("Decision Tree Classifier Classification Report:\n",
      classification_report(y_test,y_pred_dtc))

print("Random Forest Classifier Classification Report:\n",
      classification_report(y_test,y_pred_rfc))
```
