```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,ConfusionMatrixDisplay

#-----------data preprocessing------------
df=pd.read_csv("Processed_data.csv")
df
```

```python
x=df.drop('Label',axis=1)
y=df['Label']
#Train-test split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2, random_state=4)
```

```python
#feature scaling(Standarisation)
ss=StandardScaler()

x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
#training model
from sklearn.neighbors import KNeighborsClassifier
#1.using elbow method
error_rates = []

# try different K values to find the optimal value of K
for k in range(1, 21, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    pred_k = knn.predict(x_test)
    
    error = (pred_k != y_test).mean()
    error_rates.append(error)
    
k_values = list(range(1, 21, 2))
best_k = k_values[error_rates.index(min(error_rates))]
print("Best K:", best_k,"\n")
# plot graph
plt.figure(figsize=(8,5))
plt.plot(range(1,21,2), error_rates, marker='o',color='green')
plt.xlabel("K value")
plt.ylabel("Error Rate")
plt.title("Elbow Method")
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
#2.using cross validation 
k_values = range(5, 12)
scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    cv_score = cross_val_score(knn, x_train, y_train, cv=5)
    scores.append(cv_score.mean())

best_k1 = k_values[scores.index(max(scores))]
print("Best K:", best_k)
plt.plot(k_values, scores, marker='o')
plt.xlabel("K value")
plt.ylabel("Cross-Validation Accuracy")
plt.title("K vs Cross-Validation Accuracy")
plt.show()
```

```python
#According to graph(cross validation) we are getting optimal value of K as 11 
#model 1
knn = KNeighborsClassifier(n_neighbors=best_k1)
knn.fit(x_train, y_train)
```

```python
#model 2
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
#model 3
dtc=DecisionTreeClassifier(random_state=10)
dtc.fit(x_train, y_train)
```

```python
#model 4
rfc=RandomForestClassifier(n_estimators=100,random_state=10)
rfc.fit(x_train, y_train)
```

```python
#evaluation of all models and finding the best one out of all
#1.Accuracy 
y_pred_knn=knn.predict(x_test)
print("KNN Accuracy:",accuracy_score(y_test,y_pred_knn))

y_pred_lr=lr.predict(x_test)
print("Logistic Regression Accuracy:",accuracy_score(y_test,y_pred_lr))

y_pred_dtc=dtc.predict(x_test)
print("Decision Tree Classifier Accuracy:",accuracy_score(y_test,y_pred_dtc))

y_pred_rfc=rfc.predict(x_test)
print("Random Forest Classifier Accuracy:",accuracy_score(y_test,y_pred_rfc))
```

```python
#2.Confusion Matrix
cm_knn=confusion_matrix(y_test,y_pred_knn)
print("KNN Confusion matrix:\n",cm_knn)

cm_lr=confusion_matrix(y_test,y_pred_lr)
print("Logistic Regression Confusion matrix:\n",cm_lr)

cm_dt=confusion_matrix(y_test,y_pred_dtc)
print("Decision Tree Classifier Confusion matrix:\n",cm_dt)

cm_rc=confusion_matrix(y_test,y_pred_rfc)
print("Random Forest Classifier Confusion matrix:\n",cm_rc)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))

disp_knn = ConfusionMatrixDisplay(confusion_matrix=cm_knn, display_labels=["Class 0", "Class 1"])
disp_knn.plot(ax=axes[0,0], cmap="Blues", values_format='d')
axes[0,0].set_title(f"KNN (k={best_k})")

disp_lr = ConfusionMatrixDisplay(confusion_matrix=cm_lr, display_labels=["Class 0", "Class 1"])
disp_lr.plot(ax=axes[0,1], cmap="Oranges", values_format='d')
axes[0,1].set_title("Logistic Regression")

disp_dt = ConfusionMatrixDisplay(confusion_matrix=cm_dt, display_labels=["Class 0", "Class 1"])
disp_dt.plot(ax=axes[1,0], cmap="Greens", values_format='d')
axes[1,0].set_title("Decision Tree")

disp_rf = ConfusionMatrixDisplay(confusion_matrix=cm_rc, display_labels=["Class 0", "Class 1"])
disp_rf.plot(ax=axes[1,1], cmap="Purples", values_format='d')
axes[1,1].set_title("Random Forest")

plt.tight_layout()
plt.show()
```

```python
#3.Classification report
print("KNN Classification Report:\n",
      classification_report(y_test,y_pred_knn))

print("Logistic Regression Classification Report:\n",
      classification_report(y_test,y_pred_lr))

print("Decision Tree Classifier Classification Report:\n",
      classification_report(y_test,y_pred_dtc))

print("Random Forest Classifier Classification Report:\n",
      classification_report(y_test,y_pred_rfc))
```
